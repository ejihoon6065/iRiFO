```python
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈
 
# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
table = soup.find('table', { 'class': 'bbsList table' })    # <table class="table_develop3">을 찾음
table
```

```python
col_list = []
for th1 in table.find_all('tr', { 'id': 'headerNm' }):
    col_list.append(th1.text[0:2].split())
    col_list.append(th1.text[2:4].split())
    col_list.append(th1.text[4:12].split())
    col_list.append(th1.text[12:].split())
col_list
```

```python
for th1 in table.find_all('tbody', { 'id': 'contentsList' }):
    print(th1.text.split('대구(143)'))

th1.text.split('대구(143)')[1]
```

```python
loc = []
date = []
temp_max = []
temp_min = []

for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    print(tds)
    count = 0
    for td in tds:
        loc.append(tds[0].text)
        date.append(tds[1].text)
        temp_max.append(tds[2].text)
        temp_min.append(tds[3].text)
        count = count + 1
        if count == 1:
            break
        
```

```python
import pandas as pd

df = {
    'loc' : loc,
    'date' : date,
    'temp_max' : temp_max,
    'temp_min' : temp_min
}
```

```python
df = pd.DataFrame(df)
```

```python
temp_list = []
for i in col_list:
    temp_list += i
temp_list
```
